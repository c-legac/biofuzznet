```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import r2_score
```

```python
excl_runs = [0, 11, 15, 16, 17, 21, 24, 25, 26, 6, 7, 8]
runs = [i for i in range(81)]

sel_runs = [i for i in runs if i not in excl_runs]
```

```python
base_path = "/dccstor/ipc1/CAR/DREAM/Model/Test/OneCellLine/"
```

```python
val_true = pd.read_csv(f"{base_path}10/valid_data.csv", index_col=0).reset_index(drop=True)
val_pred = pd.read_csv(f"{base_path}10/valid_output_states.csv", index_col=0)

markers = [m for m in val_pred.columns if m in val_true.columns]
```

```python
val_pred[m]
```

```python
for m in markers:
    sns.scatterplot(x=val_true[m], y=val_pred[m], hue=val_true["treatment"], s =1)
    plt.title(f"{m} {r2_score(val_true[m], val_pred[m])}")
    plt.show()
```
