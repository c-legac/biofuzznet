```python
import pandas as pd
from biological_fuzzy_logic_networks.DREAM_analysis.utils import data_to_nodes_mapping
```

```python
test_cl = ['/dccstor/ipc1/CAR/DREAM/DREAMdata/AU565.csv',
 '/dccstor/ipc1/CAR/DREAM/DREAMdata/EFM19.csv',
 '/dccstor/ipc1/CAR/DREAM/DREAMdata/HCC2218.csv',
 '/dccstor/ipc1/CAR/DREAM/DREAMdata/LY2.csv',
 '/dccstor/ipc1/CAR/DREAM/DREAMdata/MACLS2.csv',
 '/dccstor/ipc1/CAR/DREAM/DREAMdata/MDAMB436.csv']
```

```python
all_cl = []
for i in range(len(test_cl)):
    data = pd.read_csv(test_cl[i])
    data = data[data["time"]==9]
    data = data.rename(columns = data_to_nodes_mapping())
    all_cl.append(data)
```

```python
all_data = pd.concat(all_cl).reset_index(drop=False)
```
