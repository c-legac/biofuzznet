```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
std_data = pd.read_csv("/dccstor/ipc1/CAR/Data/DREAM_data/Dist_fitting/t0.9_StandardScaler_dist_fitted.csv",
                      index_col=0)
```

```python
noNA = std_data[~std_data["chi_squared"].isna()]
noNA.replace(np.inf, 1e10, inplace=True)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=noNA, x="marker", y="chi_squared", hue="distribution")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(0.15,0,1,1))
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=noNA, x="distribution", y="chi_squared")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(0.2,0,1,1))
```

```python
noNA.groupby("distribution").mean().reset_index(drop=False).sort_values("chi_squared", ascending=True)
```

# Remove NaNs and inf

```python
noNA_noInf = std_data[(~std_data["chi_squared"].isna())&(std_data["chi_squared"]!=np.inf)]
noNA_noInf["chi_squared"].describe()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=noNA_noInf, x="marker", y="chi_squared", hue="distribution")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(0.15,0,1,1))
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=noNA_noInf, x="distribution", y="chi_squared")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(0.2,0,1,1))
```

```python
noNA_noInf.groupby("distribution").mean().reset_index(drop=False).sort_values("chi_squared", ascending=True)
```
